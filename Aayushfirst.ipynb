```python
#conversion of C to F and F to C
def print_menu():
  print("1. C to F")
  print("2. F to C")
  x = int(input("etr your decision:\t"))
  if x==1:
    far()
  elif x==2:
    cel()
  else:
    print("invalid")

def far():
  c1 = float(input("etr temp in C:\t"))
  f1 = (c1*9/5)+32
  print("temp in F is: {0:0.2f}".format(f1))

def cel():
  f2 = float(input("etr temp in F:\t"))
  c2 = (f2-32)*(5/9)
  print("temp in C is: {0:0.2f}".format(c2))

print_menu()
```
